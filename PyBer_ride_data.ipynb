```python
#dependencies

import numpy as np
import pandas as pd
```

```python
#load rider data into csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df.head(5)
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
```

```python
# adjust the ticks to show all the months by editing our code to look
import matplotlib.pyplot as plt
# Set x-axis and ticks locations
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]

#plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
plt.show()
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
plt.figure(figsize=(8,8))
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind="bar", color="skyblue")
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])

plt.errorbar(pyber_ride_df["Month"], pyber_ride_df["Avg. Fare ($USD)"], yerr=stdev, capsize=3, color="black",
             fmt='-'
            )
plt.xticks(rotation="horizontal")
#
plt.yticks(np.arange(0,56,5))
```
