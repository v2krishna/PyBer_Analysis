  ### Pie Chart using OOPS

```python
explode  = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
fig, ax = plt.subplots()
fig.set_size_inches(8,8) # set the size
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', 
        startangle = 80,
        explode = explode,
       counterclock=False
        )

plt.show()
```
